```python
import math
from scipy.spatial import distance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #ctrl+enter or shift+enter to run a cell on windows
```

```python
a=[1,5]
b=[1,5]
plt.figure(figsize=(14,6)).patch.set_facecolor('silver')
plt.scatter(a,b,color='red',s=80)   
plt.plot(a,b,linewidth=3,color='green')#would represent the Euclidean distance between the 2 ponts 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=[2,4]
y=[3,2]
for a,b in zip(x,y):#converts list to tuples 
    print(a,b)
```

```python
def euclidean_distance(a,b):
    return math.sqrt(sum((e1-e2)**2 for e1,e2 in zip(a,b)))
print(euclidean_distance(x,y))
```

```python
from scipy.spatial.distance import euclidean
print(euclidean(x,y))
```

```python
print(np.linalg.norm(np.array(x)-np.array(y)))
```

```python
a=[2,3]
b=[5,11]
plt.figure(figsize=(14,6)).patch.set_facecolor('silver')
plt.scatter(a,b,color='red',s=80)   
plt.plot([a[0],a[0]],b,linewidth=3,color='green')
plt.plot(a,[b[1],b[1]],linewidth=3,color='green') #For Manhattan distance
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(which='both')
plt.show()

#Computing M distance....
x_axis_distance=abs(a[0]-b[0])
y_axis_distance=abs(a[1]-b[1])
p=3
minkowski_distance=sum((x_axis_distance**p,y_axis_distance**p))**(1/p)
print("Minkowski distance=",minkowski_distance)
print(x_axis_distance+y_axis_distance)
print(distance.minkowski(a,b,p))
```

```python
def mah_distance(a,b):
    return sum(abs(m1-m2) for m1,m2 in zip(a,b))
```

```python
a=[2,3]
b=[5,11]
print(distance.cityblock(a,b)) #M distance computation using built in functions
```

```python
#Hamming distance
def calc_hamming_distance(a,b):
    return sum(abs(h1-h2) for h1,h2 in zip(a,b))/len(a)
a=[0,1]
b=[1,1]
print(calc_hamming_distance(a,b))
print(distance.hamming(a,b))
```

```python
str1='Shiva'
str2='Shabe'
lst1=list(str1)
lst2=list(str2)
print(distance.hamming(lst1,lst2)*len(str1))
```
