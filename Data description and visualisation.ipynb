```python
tc=pd.get_dummies(dataset.transmission)
tc.head()
```

```python
dataset=pd.concat([dataset,tc],axis="columns")
dataset.head()
```

```python
dataset=dataset.drop(['Manual','transmission'],axis='columns')
dataset
```
